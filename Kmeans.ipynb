# Su dung thuat toan KNN de nhan dang cac loai hoa


```
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from google.colab import files
import matplotlib.image as mpimg
```

```
anh=files.upload()
```

```
#data
iris = load_iris()
x = iris.data
y = iris.target
k=int(input('Nhap so lang gieng: '))
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(x,y)
new = np.array([[5.3,2,4.5,0.2]])
kq=knn.predict(new)[0]
print(kq)
ten_loai = iris.target_names[kq]
#nap anh
image_path= {
    "setosa": "setosa.jpg",
    "versicolor": "versicolor.jpg",
    "virginica": "virginica.jpg"
}
#Hien thi anh
imp=mpimg.imread(image_path[ten_loai])
plt.imshow(imp)
plt.axis('off')
plt.title("Loai hoa: " + ten_loai)
plt.show()
```

```
import numpy as np
from random import randint
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
points=np.array([[randint(1,10), randint(1,10) ] for i in range(100)])
plt.scatter(points[:,0], points[:,1], s=100, c='blue')
```

```
km=KMeans(n_clusters=3)
km.fit(points)
print(km.cluster_centers_)
print(km.labels_)
print(points[0], km.labels_[0])
mau=['red','cyan','green']
for i in range(len(points)):
 # plt.scatter(points[i,0], points[i,1], s=100, c=mau[km.labels_[i]])
 if km.labels_[i]==0:
  mau='red'
 elif km.labels_[i]==1:
  mau='cyan'
 else:
  mau='green'
 plt.scatter(points[i,0], points[i,1], s=100, c=mau)
```
